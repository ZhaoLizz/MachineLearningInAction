```python
import pandas as pd
```

```python
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data',header=None)
```

```python
df.tail()
```

接下来我们抽取出前100条样本，这正好是Setosa和Versicolor对应的样本，我们将Versicolor对应的数据作为类别1，Setosa对应的作为-1。对于特征，我们抽取出sepal length和petal length两维度特征，然后用散点图对数据进行可视化:

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import  pandas as pd
```

```python
y = df.iloc[0:100,4].values # 0到100行,第4列,也就是label的名字
```

```python
y = np.where(y=='Iris-setosa',-1,1)
```

```python
X = df.iloc[0:100,[0,2]].values
```

```python
# 绘制前50行
plt.scatter(X[:50,0],X[:50,1],color='red',marker='o',label='setosa')
# 绘制后50行
plt.scatter(X[50:100,0],X[50:100,1],color='blue',marker='x',label='versicolor')
plt.xlabel('petal length')
plt.ylabel('sepal length')
plt.legend(loc='upper left')
plt.show()
```

下面开始训练感知机模型,并把每一轮误分类的数目可视化出来

```python
from Perceptron import Perceptron

ppn = Perceptron(eta=0.1, n_iter=10)
```

```python
ppn.fit(X, y)
```

```python
plt.plot(range(1,len(ppn.errors_) + 1),ppn.errors_,marker='o') # X坐标点:erroes的大小(轮数),y坐标点
plt.xlabel("Epoches")
plt.ylabel("Number of misclassifications")
plt.show()
```
